# NumPy Random Arrays — Practical Notebook

This notebook demonstrates how to generate and analyze random arrays with NumPy:
1. Set up NumPy and display configuration
2. Generate uniform random arrays
3. Generate normal-distribution arrays
4. Control reproducibility with random seeds
5. Create random integer arrays
6. Apply sampling and permutation utilities
7. Inspect shapes and summary statistics
8. Visualize distributions with histograms

```python
import sys, os, warnings
warnings.filterwarnings("ignore")

# Make project root importable from notebooks/
sys.path.insert(0, os.path.join(os.getcwd(), ".."))

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import seaborn as sns

sns.set_theme(style="whitegrid", palette="tab10")
plt.rcParams["figure.dpi"] = 110

print("Python:", sys.version.split()[0])
print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
```

## 1) Set Up NumPy and Notebook Configuration

Import NumPy and configure display options for readable array output in notebook cells.

```python
np.set_printoptions(precision=4, suppress=True, linewidth=120)
pd.set_option("display.precision", 4)
pd.set_option("display.max_columns", 120)
pd.set_option("display.width", 160)

print("NumPy display options configured.")
```

## 2) Generate Uniform Random Arrays

Create 1D, 2D, and 3D arrays using uniform random values with `np.random.rand` and `np.random.random`.

```python
uniform_1d = np.random.rand(8)
uniform_2d = np.random.rand(4, 5)
uniform_3d = np.random.random((2, 3, 4))

print("1D uniform array:")
print(uniform_1d)
print("\n2D uniform array shape:", uniform_2d.shape)
print(uniform_2d)
print("\n3D uniform array shape:", uniform_3d.shape)
print(uniform_3d)
```

## 3) Generate Random Arrays from Normal Distribution

Use `np.random.randn` and `np.random.normal` to generate Gaussian-distributed values and compare parameter effects.

```python
normal_standard = np.random.randn(1_000)
normal_shifted = np.random.normal(loc=3.0, scale=0.5, size=1_000)
normal_wide = np.random.normal(loc=0.0, scale=2.0, size=1_000)

summary_df = pd.DataFrame(
    {
        "distribution": ["randn(0,1)", "normal(3,0.5)", "normal(0,2)"],
        "mean": [normal_standard.mean(), normal_shifted.mean(), normal_wide.mean()],
        "std": [normal_standard.std(), normal_shifted.std(), normal_wide.std()],
        "min": [normal_standard.min(), normal_shifted.min(), normal_wide.min()],
        "max": [normal_standard.max(), normal_shifted.max(), normal_wide.max()],
    }
)
summary_df
```

## 4) Control Reproducibility with Random Seed

Set deterministic output using `np.random.seed` and demonstrate repeated runs producing identical arrays.

```python
np.random.seed(42)
seeded_a = np.random.rand(6)

np.random.seed(42)
seeded_b = np.random.rand(6)

print("Array A:", seeded_a)
print("Array B:", seeded_b)
print("Identical:", np.array_equal(seeded_a, seeded_b))
```

## 5) Create Random Integer Arrays

Generate bounded integer arrays with `np.random.randint` for different ranges, sizes, and dimensions.

```python
randint_1d = np.random.randint(0, 10, size=12)
randint_2d = np.random.randint(100, 200, size=(3, 4))
randint_3d = np.random.randint(-5, 6, size=(2, 3, 2))

print("1D integers [0,10):", randint_1d)
print("\n2D integers [100,200):\n", randint_2d)
print("\n3D integers [-5,6): shape", randint_3d.shape)
print(randint_3d)
```

## 6) Apply Random Sampling Utilities

Use `np.random.choice`, `np.random.permutation`, and `np.random.shuffle` for sampling, permutation, and in-place shuffling.

```python
population = np.array(["A", "B", "C", "D", "E", "F"])

sample_without_replacement = np.random.choice(population, size=3, replace=False)
sample_with_probabilities = np.random.choice(population, size=10, p=[0.05, 0.1, 0.15, 0.2, 0.25, 0.25])
permuted = np.random.permutation(population)

to_shuffle = np.arange(1, 11)
np.random.shuffle(to_shuffle)

print("Population:", population)
print("Sample (no replacement):", sample_without_replacement)
print("Sample (weighted):", sample_with_probabilities)
print("Permutation:", permuted)
print("Shuffled in place:", to_shuffle)
```

## 7) Inspect Shapes, Statistics, and Value Ranges

Compute summary metrics such as min, max, mean, and standard deviation, and verify output shapes programmatically.

```python
arrays = {
    "uniform_1d": uniform_1d,
    "uniform_2d": uniform_2d,
    "normal_standard": normal_standard,
    "normal_shifted": normal_shifted,
    "randint_2d": randint_2d,
}

rows = []
for name, arr in arrays.items():
    rows.append(
        {
            "array": name,
            "shape": arr.shape,
            "dtype": str(arr.dtype),
            "min": float(np.min(arr)),
            "max": float(np.max(arr)),
            "mean": float(np.mean(arr)),
            "std": float(np.std(arr)),
        }
    )

stats_df = pd.DataFrame(rows)
stats_df
```

## 8) Visualize Random Distributions

Plot histograms for uniform and normal samples using Matplotlib to compare distribution behavior across sample sizes.

```python
sample_sizes = [100, 1_000, 10_000]
fig, axes = plt.subplots(2, 3, figsize=(16, 7), sharey="row")

for idx, n in enumerate(sample_sizes):
    u = np.random.rand(n)
    g = np.random.normal(loc=0.0, scale=1.0, size=n)

    axes[0, idx].hist(u, bins=30, edgecolor="black", alpha=0.8)
    axes[0, idx].set_title(f"Uniform [0,1), n={n}")
    axes[0, idx].set_xlabel("value")

    axes[1, idx].hist(g, bins=30, edgecolor="black", alpha=0.8)
    axes[1, idx].set_title(f"Normal (0,1), n={n}")
    axes[1, idx].set_xlabel("value")

for row in range(2):
    axes[row, 0].set_ylabel("frequency")
    for col in range(3):
        axes[row, col].yaxis.set_major_formatter(mticker.StrMethodFormatter("{x:,.0f}"))

plt.tight_layout()
plt.show()
```
